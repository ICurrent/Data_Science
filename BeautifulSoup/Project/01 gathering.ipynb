```python
from bs4 import BeautifulSoup
import os
import pandas as pd
```

```python
folder = './rt_html/'
os.listdir(folder)
```

```python
# List of dictionaries to build file by file and later convert to a DataFrame
df_list = []
folder = './rt_html'
for movie_html in os.listdir(folder):
    with open(os.path.join(folder, movie_html)) as file:
        # Your code here
        soup = BeautifulSoup(file, 'html')
        try:
            title = soup.find('h1', class_ = 'title hidden-xs').text.strip()

            audience_soup = soup.find('div', class_ = 'meter media')
            audience_score = audience_soup.find('span', class_ = 'superPageFontColor').text.strip('%')

            audience_soup = soup.find('div', class_='audience-info hidden-xs superPageFontColor')
            num_audience_ratings = audience_soup.find_all('div')[-1].text.split(':')[-1].strip().replace(',','')
            # Note: a correct implementation may take ~15 seconds to run
            
            # Append to list of dictionaries
            df_list.append({'title': title,
                            'audience_score': int(audience_score),
                            'number_of_audience_ratings': int(num_audience_ratings)})

        except Exception as e:
            print(str(title) + '\n'+ 'audience_score: '+ audience_score + '\n' +'num_audience_ratings: '+ num_audience_ratings +'\n'+ "error: " + str(e))
            print()
df = pd.DataFrame(df_list, columns = ['title', 'audience_score', 'number_of_audience_ratings'])
```

```python
df.columns = df.columns.str.capitalize()
```
